```python
import funzioniTesto  
import json
```

```python
dizionario=json.load(open('dictCompleto.json'))
```

```python
dizionario_senzaEmoji=dizionario.copy()
```

```python
dizionario_senzaEmoji['corriere'][0]['comments'][0]['text']
```

```python
len_count=dict()


for profilo,lista_post in dizionario_senzaEmoji.items():
    len_count[profilo]=dict()
    #print(lista_post)
    for idx_commenti in range(len(lista_post)):
      #print(lista_post[idx_commenti])
        
        for commenti in lista_post[idx_commenti]['comments']:
          
         commenti['text']=funzioniTesto.converti_emoji(commenti['text'])
         lingua=funzioniTesto.get_lang_text(commenti['text'])['language']

         if lingua not in len_count[profilo]:
           len_count[profilo][lingua]=0
         len_count[profilo][lingua]+=1
      #dizionario_senzaEmoji[profilo][post][commenti]
      #funzioniTesto.converti_emoji()
```

```python
for profilo, count_list in len_count.items():
  len_count[profilo]=sorted(count_list.items(), key= lambda x: x[1], reverse=True)
```

```python
len_count_post=dict()


for profilo,lista_post in dizionario_senzaEmoji.items():
   
    #print(lista_post)
    for idx_commenti in range(len(lista_post)):
      #print(lista_post[idx_commenti])
        len_count_post[lista_post[idx_commenti]['id']]=dict()
        for commenti in lista_post[idx_commenti]['comments']:
          
         commenti['text']=funzioniTesto.converti_emoji(commenti['text'])
         lingua=funzioniTesto.get_lang_text(commenti['text'])['language']
        
         
         if lingua not in len_count_post[lista_post[idx_commenti]['id']]:
           len_count_post[lista_post[idx_commenti]['id']][lingua]=0
         len_count_post[lista_post[idx_commenti]['id']][lingua]+=1
      #dizionario_senzaEmoji[profilo][post][commenti]
      #funzioniTesto.converti_emoji()
```

```python
for id, count_list in len_count_post.items():
  len_count_post[id]=sorted(count_list.items(), key= lambda x: x[1], reverse=True)
```
